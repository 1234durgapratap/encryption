```python
#dp_encryption
import getpass
def encrypt(file1,file2):
    try:
        with open(file1,'a+',encoding='utf-8') as f1:
            a1=f1.readline()
            enlist=[]
            while a1:
                a=a1.split()
                a2=[]
                l=len(a1)
                for i in a:
                    l2=len(i)
                    if l2%2==0:
                        p=i[l2//2]+i[:l2//2-1]+i[l2//2+1:]+i[l2//2-1]
                        a2.append(p)
                    else:
                        p=i[(l2-1)//2]+i[:(l2-1)//2]+i[(l2-1)//2+1:]
                        a2.append(p)
                #print(a2)
                a3=[]
                for j in a2:
                    l3=len(j)
                    if l3%2==0 or l3%3==0 or l3%5==1 :
                        p=j[::-1]
                    else:
                        p=j
                    a3.append(p)
                #print(a3)
                a4=[]
                for k in a3:
                    p=""
                    l4=len(k)
                    for l in k:
                        p+=chr(ord(l)+l4)
                    a4.append(p)
                #print(a4)
                enlist.append(a4)
                a1=f1.readline()
        
        with open(file2,'a+',encoding='utf-8') as f2:
            for e in enlist:
                for h in e:
                    f2.write(h+' ')
                f2.write("\n")
        print("-->encryption done successfully!")
    except Exception as e:
        print(e)
    
def decrypt(file1,file2):
    try:
        with open(file1,'r',encoding='utf-8') as f1:
            enlist2=[]
            a1=f1.readline()
            while a1:
                a=a1.split()
                a2=[]
                for k in a:
                    p=""
                    l4=len(k)
                    for l in k:
                        p+=chr(ord(l)-l4)
                    a2.append(p)
                #print(a2)
                a3=[]
                for j in a2:
                    l3=len(j)
                    if l3%2==0 or l3%3==0 or l3%5==1 :
                        p=j[::-1]
                    else:
                        p=j
                    a3.append(p)
                #print(a3)
                a4=[]
                l=len(a3)
                for i in a3:
                    l2=len(i)
                    if l2%2==0:
                        p=i[1:l2//2]+i[l2-1]+i[0]+i[l2//2:l2-1]
                        a4.append(p)
                    else:
                        p=i[1:(l2+1)//2]+i[0]+i[(l2+1)//2:]
                        a4.append(p)
                #print(a4)
                enlist2.append(a4)
                a1=f1.readline()
        
        with open(file2,'a+',encoding='utf-8') as f2:
            for e in enlist2:
                for h in e:
                    f2.write(h+' ')
                f2.write("\n")
        print("-->decryption done successfully!")
    except Exception as e:
        print(e)
    
#_main
password=getpass.getpass(prompt="enter password:")
if password=="53":
    file=input("enter input file name:  ")
    encrpt=input("enter output file name:   ")
    print("----------------------------")
    print("enter 1: for encryption")
    print("enter 2: for decryption")
    inp=int(input("$"))
    #funcn_call
    if inp==1:
        encrypt(file,encrpt)
    else:
        decrypt(file,encrpt)
else:
    print("service not available")
```
